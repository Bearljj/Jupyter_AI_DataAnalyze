## üß™ Step 4: Á§∫‰æãÂàÜÊûê

‰∏ãÈù¢ÊòØ‰∏Ä‰∫õÁ§∫‰æã‰ª£Á†ÅÔºåÂ±ïÁ§∫Â¶Ç‰Ωï‰ΩøÁî®Ê°ÜÊû∂

```python
# Á§∫‰æãÔºöÂàõÂª∫‰∏Ä‰∫õÊµãËØïÊï∞ÊçÆ
import polars as pl
from datetime import date, timedelta
import random

# ÁîüÊàêÁ§∫‰æãÊï∞ÊçÆ
dates = [date(2024, 1, 1) + timedelta(days=i) for i in range(100)]
products = ['‰∫ßÂìÅA', '‰∫ßÂìÅB', '‰∫ßÂìÅC']
regions = ['Âçé‰∏ú', 'ÂçéÂçó', 'ÂçéÂåó']

df_example = pl.DataFrame({
    'date': [random.choice(dates) for _ in range(1000)],
    'product': [random.choice(products) for _ in range(1000)],
    'region': [random.choice(regions) for _ in range(1000)],
    'premium': [random.randint(1000, 50000) for _ in range(1000)],
    'claims': [random.randint(0, 30000) for _ in range(1000)]
})

print("Á§∫‰æãÊï∞ÊçÆÔºö")
print(df_example.head())
print(f"\nÊï∞ÊçÆÈáè: {len(df_example)} Ë°å")
```

### Âü∫Á°ÄÂàÜÊûêÁ§∫‰æã

```python
# Êåâ‰∫ßÂìÅÂàÜÁªÑËÅöÂêà
by_product = df_example.group_by('product').agg([
    pl.col('premium').sum().alias('total_premium'),
    pl.col('claims').sum().alias('total_claims'),
    pl.count().alias('count')
]).with_columns(
    (pl.col('total_claims') / pl.col('total_premium')).alias('loss_ratio')
).sort('total_premium', descending=True)

print("Êåâ‰∫ßÂìÅËÅöÂêàÔºö")
print(by_product)
```

### ÂèØËßÜÂåñÁ§∫‰æã

```python
# ÂàõÂª∫Êù°ÂΩ¢Âõæ
fig = px.bar(
    by_product.to_pandas(),
    x='product',
    y='total_premium',
    title='ÂêÑ‰∫ßÂìÅ‰øùË¥πÊÄªÈ¢ù',
    labels={'total_premium': '‰øùË¥πÊÄªÈ¢ù', 'product': '‰∫ßÂìÅ'}
)

fig.update_layout(height=400)
fig.show()
```
