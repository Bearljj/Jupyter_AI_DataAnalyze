## ğŸ¤– Step 6: ç²˜è´´ AI é€»è¾‘

```python
import panel as pn
import polars as pl
import plotly.express as px
from src.dashboard import PanelDashboardBuilder
from src.utils import print_markdown_table

# ==========================================
# 1. åˆå§‹åŒ–ä»ªè¡¨ç›˜
# ==========================================
# æ ¸å¿ƒç»´åº¦
dimensions = ['å¢ƒå†…å¢ƒå¤–', 'å½’å±æœºæ„', 'ä¸šåŠ¡å¹´åº¦']

dashboard = PanelDashboardBuilder.from_data(
    df_df,
    dimensions=dimensions,
    title="å¢ƒå†…å¤–ä¸šåŠ¡ä¿è´¹å…¨æ™¯åˆ†æï¼ˆåˆ†å‡º vs è‡ªç•™ï¼‰"
)

# ==========================================
# 2. å®šä¹‰åˆ†æé€»è¾‘
# ==========================================
@pn.depends(*dashboard.widgets.values())
def update_dashboard(*args):
    """
    [REPORT_METADATA]
    ### 1. ä¸šåŠ¡èƒŒæ™¯
    å…¨æ™¯å±•ç¤ºä¸šåŠ¡çš„èµ„é‡‘æµå‘ç»“æ„ï¼Œå¯¹æ¯”å¢ƒå†…ä¸å¢ƒå¤–ä¸šåŠ¡åœ¨â€œè‡ªç•™â€ä¸â€œåˆ†å‡ºâ€ä¸Šçš„ç­–ç•¥å·®å¼‚ï¼Œç›‘æ§å„åˆ†ä¿æ–¹å¼çš„æˆæœ¬è´¡çŒ®ã€‚
    
    ### 2. åˆ†æé€»è¾‘
    * **æ ¸å¿ƒå¤„ç†**ï¼šå°†æ•°æ®æŒ‰â€œèšåˆç»´åº¦â€å’Œâ€œå¢ƒå†…å¢ƒå¤–â€åˆ†ç»„åï¼Œå¯¹7ç§è´¹ç”¨ç±»å‹è¿›è¡Œ**é€†é€è§†(Unpivot)**ï¼Œå½¢æˆå †å å›¾æ•°æ®æºã€‚
    * **è´¹ç”¨æ„æˆ**ï¼š
        * **è‡ªç•™**ï¼šè‡ªç•™ä¿è´¹ + é™„åŠ è‡ªç•™ä¿è´¹
        * **åˆ†å‡º**ï¼šåè®® + æˆæ•° + æº¢é¢ + OpenCover + ä¸´åˆ†
    * **å¯è§†åŒ–**ï¼šåŒå±‚Xè½´ï¼ˆç»´åº¦+å¢ƒå†…å¢ƒå¤–ï¼‰åˆ†ç»„ï¼Œé¢œè‰²æ˜ å°„è´¹ç”¨ç±»å‹ã€‚
    [/REPORT_METADATA]
    """
    # --- 1. è·å–è¾“å…¥ ---
    filters = dashboard.data_values
    agg_dim = dashboard.widgets['_aggregation_dimension'].value
    
    # å¼ºåˆ¶å°† 'å¢ƒå†…å¢ƒå¤–' ä»è¿‡æ»¤ç»´åº¦ä¸­ç§»é™¤ï¼ˆå› ä¸ºå®ƒå°†ä½œä¸ºå¯¹æ¯”ç»´åº¦å›ºå®šåœ¨å›¾è¡¨ä¸­ï¼‰
    # ä½†å¦‚æœç”¨æˆ·åœ¨æ§ä»¶é‡Œé€‰äº†'å…¨é€‰'æˆ–ç‰¹å®šå€¼ï¼Œæˆ‘ä»¬éœ€è¦å°Šé‡è¿™ä¸ªè¿‡æ»¤
    # ç­–ç•¥ï¼šå³ä½¿ä½œä¸ºå›¾è¡¨åˆ†ç»„ï¼Œä¹Ÿåº”å…ˆå“åº”è¿‡æ»¤
    
    # --- 2. æ•°æ®è¿‡æ»¤ ---
    filtered = df_df
    for dim, val in filters.items():
        if isinstance(val, list):
            if 'å…¨é€‰' not in val:
                filtered = filtered.filter(pl.col(dim).is_in(val))
        elif val != 'å…¨é€‰':
            filtered = filtered.filter(pl.col(dim) == val)
            
    # --- 3. èšåˆè®¡ç®— (ç¬¬ä¸€æ­¥ï¼šå®½è¡¨) ---
    # å®šä¹‰7ä¸ªæŒ‡æ ‡çš„èšåˆåˆ—è¡¨
    metrics = [
        # è‡ªç•™ç«¯
        pl.col('è‡ªç•™ä¿è´¹').fill_null(0).sum(),
        pl.col('é™„åŠ è‡ªç•™ä¿è´¹').fill_null(0).sum(),
        # åˆ†å‡ºç«¯
        pl.col('åè®®åˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('æˆæ•°åˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('æº¢é¢åˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('OPENCOVERåˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('ä¸´åˆ†åˆ†ä¿è´¹').fill_null(0).sum()
    ]
    
    # å¿…é¡»åŒæ—¶æŒ‰ [é€‰å®šç»´åº¦, å¢ƒå†…å¢ƒå¤–] åˆ†ç»„
    group_cols = [agg_dim, 'å¢ƒå†…å¢ƒå¤–']
    # å¦‚æœé€‰å®šç»´åº¦å°±æ˜¯å¢ƒå†…å¢ƒå¤–ï¼Œå»é‡
    if agg_dim == 'å¢ƒå†…å¢ƒå¤–':
        group_cols = ['å¢ƒå†…å¢ƒå¤–']
        
    agg_df = filtered.group_by(group_cols).agg(metrics)
    
    # --- 4. é€†é€è§† (Unpivot) (ç¬¬äºŒæ­¥ï¼šè½¬é•¿è¡¨ä»¥ä¾¿å †å ) ---
    # ç¡®å®šè¦è½¬æ¢çš„åˆ—ï¼ˆæ•°å€¼åˆ—ï¼‰
    value_vars = [
        'è‡ªç•™ä¿è´¹', 'é™„åŠ è‡ªç•™ä¿è´¹', 
        'åè®®åˆ†ä¿è´¹', 'æˆæ•°åˆ†ä¿è´¹', 'æº¢é¢åˆ†ä¿è´¹', 'OPENCOVERåˆ†ä¿è´¹', 'ä¸´åˆ†åˆ†ä¿è´¹'
    ]
    
    # ä½¿ç”¨ melt (unpivot)
    melted_df = agg_df.melt(
        id_vars=group_cols,
        value_vars=value_vars,
        variable_name='è´¹ç”¨ç±»å‹',
        value_name='é‡‘é¢'
    )
    
    # --- 5. å•ä½è½¬æ¢ ---
    max_val = melted_df['é‡‘é¢'].max()
    if max_val and max_val > 1_000_000:
        plot_df = melted_df.with_columns(
            (pl.col('é‡‘é¢') / 10000).alias('é‡‘é¢(ä¸‡å…ƒ)')
        )
        y_col = 'é‡‘é¢(ä¸‡å…ƒ)'
        title_suffix = 'ï¼ˆå•ä½ï¼šä¸‡å…ƒï¼‰'
    else:
        plot_df = melted_df
        y_col = 'é‡‘é¢'
        title_suffix = 'ï¼ˆå•ä½ï¼šå…ƒï¼‰'

    # --- 6. æ’åºä¼˜åŒ– ---
    # ä¸ºäº†è®©å †å é¡ºåºå›ºå®šï¼ˆè‡ªç•™åœ¨çº¿é¢ï¼Œåˆ†å‡ºåœ¨ä¸Šé¢ï¼‰ï¼Œæˆ‘ä»¬éœ€è¦å®šä¹‰é¡ºåº
    # è¿™é‡Œé€šè¿‡ Pandas çš„ Categorical å®ç°ï¼Œæˆ–è€… Plotly category_orders
    category_order = [
        'è‡ªç•™ä¿è´¹', 'é™„åŠ è‡ªç•™ä¿è´¹', 
        'åè®®åˆ†ä¿è´¹', 'æˆæ•°åˆ†ä¿è´¹', 'æº¢é¢åˆ†ä¿è´¹', 'OPENCOVERåˆ†ä¿è´¹', 'ä¸´åˆ†åˆ†ä¿è´¹'
    ]
    
    # --- 7. è¾“å‡ºè¡¨æ ¼ ---
    print(f"## æŒ‰ {agg_dim} å¢ƒå†…å¤–ä¿è´¹æ„æˆåˆ†æ\n")
    
    # å±•ç¤ºåŸå§‹èšåˆæ•°æ®ï¼ˆå®½è¡¨ï¼‰ä»¥ä¾¿æŸ¥é˜…æ€»æ•°
    display_cols = group_cols + ['è‡ªç•™ä¿è´¹', 'åè®®åˆ†ä¿è´¹', 'ä¸´åˆ†åˆ†ä¿è´¹'] # é€‰å‡ ä¸ªå…³é”®åˆ—å±•ç¤º
    print_markdown_table(agg_df.select(display_cols).head(10))
    
    # --- 8. å¯è§†åŒ– ---
    # æ„å»º X è½´ï¼šå¦‚æœæ˜¯ä¸åŒç»´åº¦ï¼Œä½¿ç”¨åˆ—è¡¨å®ç°åˆ†ç»„æ•ˆæœ
    x_axis = group_cols if len(group_cols) > 1 else group_cols[0]
    
    fig = px.bar(
        plot_df.to_pandas(),
        x=x_axis,
        y=y_col,
        color='è´¹ç”¨ç±»å‹',
        title=f'å¢ƒå†…å¤–ä¸šåŠ¡ä¿è´¹æ„æˆå †å å›¾ {title_suffix}',
        category_orders={'è´¹ç”¨ç±»å‹': category_order}, # å›ºå®šå †å é¡ºåº
        text_auto='.2s'
    )
    
    fig.update_layout(
        height=600,
        autosize=True,
        barmode='stack', # æ˜ç¡®æŒ‡å®šå †å 
        xaxis_title=" / ".join(group_cols),
        yaxis_title=y_col,
        legend_title='ä¿è´¹æ„æˆ'
    )
    
    return fig

# ==========================================
# 3. å¯åŠ¨
# ==========================================
dashboard.set_update_function(update_dashboard)
dashboard.show()
```
