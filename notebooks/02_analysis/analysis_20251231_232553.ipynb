## ğŸ¤– Step 6: ç²˜è´´ AI é€»è¾‘

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import polars as pl
import panel as pn
from src.utils import print_markdown_table

# âš ï¸ ç¡®ä¿ dashboard å¯¹è±¡å·²å­˜åœ¨
# dashboard = PanelDashboardBuilder.from_data(...)

@pn.depends(*dashboard.widgets.values())
def update_dashboard(*args):
    """
    [REPORT_METADATA]
    ### 1. ä¸šåŠ¡èƒŒæ™¯
    å¯¹æº¢é¢åˆçº¦è¿›è¡Œâ€œæŠ•å…¥(ä¿è´¹)-äº§å‡º(æŸç›Š)â€çš„å…¨æ™¯åˆ†æã€‚åŒºåˆ†â€œè¶…æ³•å®šè‡ªç•™â€å¤§é¢ä¸šåŠ¡ä¸â€œé£é™©åˆ†æ•£â€å¸¸è§„ä¸šåŠ¡ï¼Œè¯„ä¼°åˆ†ä¿ç»“æ„çš„åˆç†æ€§ã€‚

    ### 2. åˆ†æé€»è¾‘
    - **åˆ†ç±»æ ‡å‡†**: (è‡ªç•™+é™„åŠ +æº¢é¢)ä¿é¢ â‰¥ 4.5äº¿ æ ‡è®°ä¸º 'è¶…æ³•å®šè‡ªç•™'ï¼Œå¦åˆ™ä¸º 'é£é™©åˆ†æ•£'ã€‚
    - **å›¾è¡¨ç»“æ„**: 
        - ä¸Šå›¾ (ä¿è´¹): æº¢é¢åˆ†ä¿è´¹ (å †å æŸ±çŠ¶å›¾)ï¼Œå±•ç¤ºåˆ†å‡ºè§„æ¨¡æ„æˆã€‚
        - ä¸‹å›¾ (æŸç›Š): å†ä¿æ‰¿ä¿åˆ©æ¶¦ (åˆ†ç»„æŸ±çŠ¶å›¾)ï¼Œå±•ç¤ºåˆçº¦ç›ˆäºç»“æœã€‚
    - **å•ä½**: ç»Ÿä¸€è½¬æ¢ä¸ºä¸‡å…ƒã€‚
    [/REPORT_METADATA]
    """
    # 1. ğŸ›¡ï¸ ç‰©ç†éš”ç¦»
    filters = dashboard.data_values
    agg_dim = dashboard.widgets['_aggregation_dimension'].value
    
    # --- A. åŠ¨æ€è¿‡æ»¤ ---
    filtered = df_df
    for col, val in filters.items():
        if isinstance(val, list):
            if 'å…¨é€‰' not in val:
                filtered = filtered.filter(pl.col(col).is_in(val))
        elif val != 'å…¨é€‰':
            filtered = filtered.filter(pl.col(col) == val)
            
    # --- B. æ ¸å¿ƒé€»è¾‘ï¼šåˆ†ç±»ä¸è®¡ç®— ---
    THRESHOLD = 450_000_000
    
    # é¢„è®¡ç®—ï¼šåˆ†ç±»ã€ä¿è´¹ã€æŸç›Š
    calculated = filtered.select([
        pl.col(agg_dim),
        
        # 1. åˆ†ç±»
        pl.when(
            (pl.col('è‡ªç•™ä¿é¢').fill_null(0) + 
             pl.col('é™„åŠ è‡ªç•™ä¿é¢').fill_null(0) + 
             pl.col('æº¢é¢åˆ†ä¿é¢').fill_null(0)) >= THRESHOLD
        ).then(pl.lit('è¶…æ³•å®šè‡ªç•™'))
         .otherwise(pl.lit('é£é™©åˆ†æ•£'))
         .alias('ä¸šåŠ¡åˆ†ç±»'),
         
        # 2. ä¿è´¹ (åˆ†å‡ºè§„æ¨¡)
        pl.col('æº¢é¢åˆ†ä¿è´¹').fill_null(0).alias('ä¿è´¹_raw'),
        
        # 3. æŸç›Š (å†ä¿åˆ©æ¶¦)
        (pl.col('æº¢é¢åˆ†ä¿è´¹').fill_null(0) - 
         (pl.col('æº¢é¢å·²å†³').fill_null(0) + pl.col('æº¢é¢æœªå†³').fill_null(0)) - 
         pl.col('æº¢é¢æ‘Šå›è´¹ç”¨').fill_null(0)
        ).alias('æŸç›Š_raw')
    ])
    
    # --- C. èšåˆåˆ†æ ---
    analysis = calculated.group_by([agg_dim, 'ä¸šåŠ¡åˆ†ç±»']).agg([
        pl.col('ä¿è´¹_raw').sum().alias('ä¿è´¹é¢'),
        pl.col('æŸç›Š_raw').sum().alias('æŸç›Šé¢'),
        pl.len().alias('ä¿å•æ•°')
    ]).sort(agg_dim)
    
    # --- D. å•ä½è½¬æ¢ (ä¸‡å…ƒ) ---
    unit = "ä¸‡å…ƒ"
    div = 10_000
    
    analysis = analysis.with_columns([
        (pl.col('ä¿è´¹é¢') / div).alias('ä¿è´¹'),
        (pl.col('æŸç›Šé¢') / div).alias('æŸç›Š')
    ])
    
    # è½¬ Pandas æ‹†åˆ†æ•°æ®ï¼Œæ–¹ä¾¿ç”¨ go.Bar ç»˜å›¾
    df_pd = analysis.to_pandas()
    
    # æ‹†åˆ†ä¸¤ç±»æ•°æ®ï¼Œç¡®ä¿å¯¹é½ X è½´
    df_super = df_pd[df_pd['ä¸šåŠ¡åˆ†ç±»'] == 'è¶…æ³•å®šè‡ªç•™']
    df_normal = df_pd[df_pd['ä¸šåŠ¡åˆ†ç±»'] == 'é£é™©åˆ†æ•£']
    
    # --- E. å¯è§†åŒ– (åŒå­å›¾) ---
    fig = make_subplots(
        rows=2, cols=1, 
        shared_xaxes=True,
        vertical_spacing=0.1,
        subplot_titles=(f"ğŸ“ˆ åˆ†å‡ºä¿è´¹è§„æ¨¡ ({unit}) - å †å å›¾", f"ğŸ“‰ å†ä¿æ‰¿ä¿åˆ©æ¶¦ ({unit}) - åˆ†ç»„å¯¹æ¯”")
    )

    # é¢œè‰²å®šä¹‰
    COLOR_SUPER = '#EF553B'  # çº¢
    COLOR_NORMAL = '#636EFA' # è“

    # === Row 1: ä¿è´¹ (å †å ) ===
    fig.add_trace(
        go.Bar(x=df_super[agg_dim], y=df_super['ä¿è´¹'], name='è¶…æ³•å®š-ä¿è´¹', 
               marker_color=COLOR_SUPER, opacity=0.6, legendgroup='super'),
        row=1, col=1
    )
    fig.add_trace(
        go.Bar(x=df_normal[agg_dim], y=df_normal['ä¿è´¹'], name='é£é™©åˆ†æ•£-ä¿è´¹', 
               marker_color=COLOR_NORMAL, opacity=0.6, legendgroup='normal'),
        row=1, col=1
    )

    # === Row 2: æŸç›Š (åˆ†ç»„) ===
    fig.add_trace(
        go.Bar(x=df_super[agg_dim], y=df_super['æŸç›Š'], name='è¶…æ³•å®š-æŸç›Š', 
               marker_color=COLOR_SUPER, legendgroup='super', showlegend=False), # å…±äº«å›¾ä¾‹
        row=2, col=1
    )
    fig.add_trace(
        go.Bar(x=df_normal[agg_dim], y=df_normal['æŸç›Š'], name='é£é™©åˆ†æ•£-æŸç›Š', 
               marker_color=COLOR_NORMAL, legendgroup='normal', showlegend=False),
        row=2, col=1
    )
    
    # æ·»åŠ  0 çº¿å‚è€ƒ
    fig.add_hline(y=0, line_dash="dash", line_color="gray", row=2, col=1)

    # å¸ƒå±€è®¾ç½®
    fig.update_layout(
        height=700, # åŠ é«˜ä¸€ç‚¹
        autosize=True,
        barmode='relative', # Row 1 å †å  (é»˜è®¤è¡Œä¸ºé…åˆ opacity)
        hovermode="x unified",
        margin=dict(l=40, r=40, t=60, b=40),
        legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1)
    )
    
    # å¼ºåˆ¶ Row 2 ä¸ºåˆ†ç»„æ¨¡å¼ (Grouped)
    # Plotly å°æŠ€å·§ï¼šå…¨å±€ barmode ä¼šå½±å“æ‰€æœ‰ traceï¼Œè¿™é‡Œæˆ‘ä»¬éœ€è¦æ··åˆæ¨¡å¼
    # å®é™…ä¸Š Plotly ä¸æ”¯æŒåŒä¸€ä¸ª fig æ··åˆ 'stack' å’Œ 'group'ã€‚
    # è§£å†³æ–¹æ¡ˆï¼šå…¨éƒ¨è®¾ä¸º 'group'ï¼Œä½† Row 1 æ‰‹åŠ¨è°ƒæ•´åç§»ï¼Ÿæˆ–è€…æ¥å— Row 1 ä¹Ÿæ˜¯ groupã€‚
    # ä¸ºäº†æ¸…æ™°ï¼Œæˆ‘ä»¬ç»Ÿä¸€ä½¿ç”¨ 'group' (åˆ†ç»„å¯¹æ¯”)ï¼Œè¿™æ ·ä¸Šä¸‹å¯¹åº”æ›´æ•´é½ã€‚
    fig.update_layout(barmode='group') 

    # --- F. Markdown æŠ¥å‘Š ---
    print(f"## ğŸ“Š æº¢é¢åˆçº¦å…¨æ™¯æŠ¥å‘Š ({agg_dim})\n")
    
    # æ±‡æ€»è®¡ç®—
    total_prem = analysis['ä¿è´¹é¢'].sum() / div
    total_profit = analysis['æŸç›Šé¢'].sum() / div
    profit_margin = (total_profit / total_prem) if total_prem > 0 else 0
    
    print(f"- **æ€»åˆ†å‡ºä¿è´¹**: {total_prem:,.2f} {unit}")
    print(f"- **æ€»å†ä¿æŸç›Š**: {total_profit:,.2f} {unit} (åˆ©æ¶¦ç‡: {profit_margin:.1%})")
    print("> ğŸ’¡ **å›¾è¡¨è§£è¯»**: \n> - **ä¸Šå›¾**: çœ‹åˆ†ç»™å†ä¿äººå¤šå°‘é’±ï¼ˆæŸ±å­è¶Šé«˜ï¼Œåˆ†å‡ºè¶Šå¤šï¼‰ã€‚\n> - **ä¸‹å›¾**: çœ‹å†ä¿äººèµšäº†è¿˜æ˜¯äºäº†ï¼ˆ0è½´ä»¥ä¸‹è¡¨ç¤ºæˆ‘å¸æˆåŠŸæ‘Šå›è¶…é¢æŸå¤±ï¼‰ã€‚\n")

    print(f"### è¯¦ç»†æ•°æ®é€è§† ({unit})")
    
    # æ„é€ é€è§†è¡¨
    display_df = analysis.pivot(
        index=agg_dim, 
        on='ä¸šåŠ¡åˆ†ç±»', 
        values=['ä¿è´¹', 'æŸç›Š'],
        aggregate_function='sum'
    ).sort(agg_dim)
    
    # ç¨å¾®ç¾åŒ–åˆ—å (å¦‚æœ Polars ç‰ˆæœ¬æ”¯æŒ)
    try:
        cols = sorted(display_df.columns)
        display_df = display_df.select([agg_dim] + [c for c in cols if c != agg_dim])
    except:
        pass

    print_markdown_table(display_df.head(10))
    
    return fig

dashboard.set_update_function(update_dashboard)
dashboard.show()
```
