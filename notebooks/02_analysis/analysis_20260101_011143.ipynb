## ğŸ¤– Step 6: ç²˜è´´ AI é€»è¾‘

```python
import panel as pn
import polars as pl
import plotly.express as px
from src.dashboard import PanelDashboardBuilder
from src.utils import print_markdown_table

# ==========================================
# 1. åˆå§‹åŒ–ä»ªè¡¨ç›˜
# ==========================================
dimensions = ['å¢ƒå†…å¢ƒå¤–', 'å½’å±æœºæ„', 'ä¸šåŠ¡å¹´åº¦']

dashboard = PanelDashboardBuilder.from_data(
    df_df,
    dimensions=dimensions,
    title="å¢ƒå†…å¤–ä¸šåŠ¡ä¿è´¹å…¨æ™¯åˆ†æï¼ˆä¸‡å…ƒç‰ˆï¼‰"
)

# ==========================================
# 2. å®šä¹‰åˆ†æé€»è¾‘
# ==========================================
@pn.depends(*dashboard.widgets.values())
def update_dashboard(*args):
    """
    [REPORT_METADATA]
    ### 1. ä¸šåŠ¡èƒŒæ™¯
    æœ¬æŠ¥å‘Šé‡‡ç”¨åˆ†é¢å¯¹æ¯”è§†è§’ï¼Œåˆ†æå¢ƒå†…ä¸å¢ƒå¤–ä¸šåŠ¡çš„ä¿è´¹æ„æˆå·®å¼‚ã€‚æ‰€æœ‰é‡‘é¢æŒ‡æ ‡å‡ç»Ÿä¸€é‡‡ç”¨**â€œä¸‡å…ƒâ€**ä¸ºå•ä½ï¼Œä»¥ä¾¿äºå®è§‚é˜…è¯»ã€‚
    
    ### 2. åˆ†æé€»è¾‘
    * **å¯è§†åŒ–ç­–ç•¥**ï¼šåˆ†é¢å †å æŸ±çŠ¶å›¾ (Faceted Stacked Bar)ã€‚
    * **æ•°æ®å¤„ç†**ï¼š
        1. èšåˆï¼šæŒ‰ `[èšåˆç»´åº¦, å¢ƒå†…å¢ƒå¤–]` ç»Ÿè®¡ 7 ç§è´¹ç”¨ã€‚
        2. **å•ä½è½¬æ¢**ï¼šæ‰€æœ‰é‡‘é¢å­—æ®µé™¤ä»¥ 10,000ã€‚
        3. æ’åºï¼šæŒ‰æ—¶é—´è½´æ’åºã€‚
        4. é€†é€è§†ï¼šä½¿ç”¨ `unpivot` è½¬ä¸ºé•¿è¡¨ã€‚
    * **æŒ‡æ ‡æ„æˆ**ï¼šè‡ªç•™ç«¯ vs åˆ†å‡ºç«¯ã€‚
    [/REPORT_METADATA]
    """
    # --- 1. è·å–è¾“å…¥ ---
    filters = dashboard.data_values
    agg_dim = dashboard.widgets['_aggregation_dimension'].value
    
    # --- 2. æ•°æ®è¿‡æ»¤ ---
    filtered = df_df
    for dim, val in filters.items():
        if isinstance(val, list):
            if 'å…¨é€‰' not in val:
                filtered = filtered.filter(pl.col(dim).is_in(val))
        elif val != 'å…¨é€‰':
            filtered = filtered.filter(pl.col(dim) == val)
            
    # --- 3. èšåˆè®¡ç®— ---
    # åœ¨èšåˆé˜¶æ®µç›´æ¥è½¬æ¢ä¸ºä¸‡å…ƒï¼Ÿæˆ–è€…èšåˆåè½¬æ¢ã€‚
    # ä¸ºäº†ç²¾åº¦ï¼Œå»ºè®®å…ˆèšåˆåè½¬æ¢ã€‚
    metrics = [
        pl.col('è‡ªç•™ä¿è´¹').fill_null(0).sum(),
        pl.col('é™„åŠ è‡ªç•™ä¿è´¹').fill_null(0).sum(),
        pl.col('åè®®åˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('æˆæ•°åˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('æº¢é¢åˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('OPENCOVERåˆ†ä¿è´¹').fill_null(0).sum(),
        pl.col('ä¸´åˆ†åˆ†ä¿è´¹').fill_null(0).sum()
    ]
    
    # ç¡®ä¿åˆ†ç»„åŒ…å« 'å¢ƒå†…å¢ƒå¤–'
    group_cols = [agg_dim]
    if 'å¢ƒå†…å¢ƒå¤–' not in group_cols:
        group_cols.append('å¢ƒå†…å¢ƒå¤–')
        
    agg_df = filtered.group_by(group_cols).agg(metrics)
    
    # --- 4. é€†é€è§† (Unpivot) ---
    value_vars = [
        'è‡ªç•™ä¿è´¹', 'é™„åŠ è‡ªç•™ä¿è´¹', 
        'åè®®åˆ†ä¿è´¹', 'æˆæ•°åˆ†ä¿è´¹', 'æº¢é¢åˆ†ä¿è´¹', 'OPENCOVERåˆ†ä¿è´¹', 'ä¸´åˆ†åˆ†ä¿è´¹'
    ]
    
    melted_df = agg_df.unpivot(
        index=group_cols,
        on=value_vars,
        variable_name='è´¹ç”¨ç±»å‹',
        value_name='é‡‘é¢'
    )
    
    # --- 5. å¼ºåˆ¶å•ä½è½¬æ¢ (ä¸‡å…ƒ) ---
    plot_df = melted_df.with_columns(
        (pl.col('é‡‘é¢') / 10000).alias('é‡‘é¢(ä¸‡å…ƒ)')
    )
    y_col = 'é‡‘é¢(ä¸‡å…ƒ)'
    title_suffix = 'ï¼ˆå•ä½ï¼šä¸‡å…ƒï¼‰'

    # --- 6. æ’åº (ä¿æŒæ—¶é—´è½´é¡ºåº) ---
    plot_df = plot_df.sort(agg_dim)
    
    # å®šä¹‰å †å é¡ºåº
    category_order = {
        'è´¹ç”¨ç±»å‹': ['è‡ªç•™ä¿è´¹', 'é™„åŠ è‡ªç•™ä¿è´¹', 'åè®®åˆ†ä¿è´¹', 'æˆæ•°åˆ†ä¿è´¹', 'æº¢é¢åˆ†ä¿è´¹', 'OPENCOVERåˆ†ä¿è´¹', 'ä¸´åˆ†åˆ†ä¿è´¹']
    }
    
    # --- 7. è¾“å‡ºè¡¨æ ¼ ---
    print(f"## æŒ‰ {agg_dim} å¢ƒå†…å¤–åˆ†é¢åˆ†æ (ä¸‡å…ƒ)\n")
    
    # è¡¨æ ¼å±•ç¤ºä¹Ÿè½¬æ¢ä¸ºä¸‡å…ƒä»¥ä¾¿ä¸€è‡´
    display_df = agg_df.sort(agg_dim).select(
        group_cols + [
            (pl.col(c) / 10000).alias(f"{c}(ä¸‡å…ƒ)") for c in value_vars
        ]
    )
    # æ ¼å¼åŒ–ä¿ç•™2ä½å°æ•°
    display_df = display_df.with_columns([
        pl.col(c).map_elements(lambda x: f"{x:,.2f}", return_dtype=pl.Utf8) 
        for c in display_df.columns if 'ä¸‡å…ƒ' in c
    ])
    
    print_markdown_table(display_df.head(5))
    
    # --- 8. å¯è§†åŒ– ---
    facet_col = 'å¢ƒå†…å¢ƒå¤–' if agg_dim != 'å¢ƒå†…å¢ƒå¤–' else None
    
    fig = px.bar(
        plot_df.to_pandas(),
        x=agg_dim,
        y=y_col,
        color='è´¹ç”¨ç±»å‹',
        facet_col=facet_col,
        title=f'å¢ƒå†…å¤–ä¸šåŠ¡ä¿è´¹æ„æˆå¯¹æ¯” {title_suffix}',
        category_orders=category_order,
        text_auto='.2f' # ä¿ç•™2ä½å°æ•°
    )
    
    fig.update_layout(
        height=600,
        autosize=True,
        barmode='stack',
        legend_title='è´¹ç”¨æ„æˆ'
    )
    
    fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
    
    return fig

# ==========================================
# 3. å¯åŠ¨
# ==========================================
dashboard.set_update_function(update_dashboard)
dashboard.show()
```
