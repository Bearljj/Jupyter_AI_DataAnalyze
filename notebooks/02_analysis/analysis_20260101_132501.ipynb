## ğŸ¤– Step 6: ç²˜è´´ AI é€»è¾‘

```python
import polars as pl
import panel as pn

# å°è¯•å¯¼å…¥æ ‡å‡†å·¥å…·
try:
    from src.utils import print_markdown_table
except ImportError:
    def print_markdown_table(df):
        try:
            print(df.to_pandas().to_markdown(index=False, floatfmt=".2f"))
        except:
            print(df)

@pn.depends(*dashboard.widgets.values())
def update_dashboard(*args):
    """
    [REPORT_METADATA]
    ### 1. ä¸šåŠ¡èƒŒæ™¯
    å•ä¸€å¤§æ¡ˆ (å±±è¥¿æ±¾é£æ½‡æ²³) æç«¯å‹åŠ›æµ‹è¯• (ä¿®å¤ç‰ˆ v6.1).
    
    ### 2. ä¿®å¤è¯´æ˜
    - **Bugä¿®å¤**: è¡¥å…¨äº† 'REINS_EXPENSE' å’Œ 'FIXED_COMM' å˜é‡å®šä¹‰ã€‚
    - **æµ‹è¯•åœºæ™¯**: 2023å¹´å¢åŠ  1.02äº¿ èµ”æ¬¾å†²å‡» (æ€»æŸ1.4äº¿åœºæ™¯)ã€‚
    [/REPORT_METADATA]
    """
    
    # -------------------------------------------------------------------------
    # 1. å‚æ•°è®¾ç½® (å…¨å±€)
    # -------------------------------------------------------------------------
    TREND_RATE = 0.05
    REINS_EXPENSE = 0.05    # âœ… è¡¥å…¨: å†ä¿äººè´¹ç”¨ 5%
    FIXED_COMM = 0.325      # âœ… è¡¥å…¨: å›ºå®šæ‰‹ç»­è´¹ 32.5%
    
    LPC_TRIGGER, LPC_CAP, LPC_SHARE = 0.67, 1.20, 0.30
    LPC_RANGE = LPC_CAP - LPC_TRIGGER
    
    # å‹åŠ›å‚æ•°
    STRESS_YEAR = 2023
    STRESS_CEDED_LOSS = 119_295_300  # 1.19äº¿ (åŸºäº 1.4äº¿æ€»æŸ * 85.21% åˆ†å‡º)
    CURRENT_BOOKED_LOSS = 17_042_200 # 1704ä¸‡ (è´¦é¢å·²ä½“ç°)
    DELTA_LOSS = STRESS_CEDED_LOSS - CURRENT_BOOKED_LOSS # +1.02äº¿
    
    # -------------------------------------------------------------------------
    # 2. æ•°æ®å‡†å¤‡
    # -------------------------------------------------------------------------
    filters = dashboard.data_values
    filtered_df = df_df
    for col, val in filters.items():
        if val is None or val == 'å…¨é€‰': continue
        if isinstance(val, list) and ('å…¨é€‰' in val or not val): continue
        if isinstance(val, list):
            filtered_df = filtered_df.filter(pl.col(col).is_in(val))
        else:
            filtered_df = filtered_df.filter(pl.col(col) == val)
            
    # é”å®šæº¢é¢åˆçº¦
    surplus_cols = ["æº¢é¢åˆ†ä¿è´¹", "æº¢é¢å·²å†³", "æº¢é¢æœªå†³", "æº¢é¢æ‘Šå›è´¹ç”¨"]
    mask = pl.lit(False)
    for col in surplus_cols: mask = mask | (pl.col(col) != 0)
    filtered_df = filtered_df.filter(mask)
    if filtered_df.height == 0: return None

    # èšåˆå†å²
    history_df = filtered_df.group_by("ä¸šåŠ¡å¹´åº¦").agg([
        pl.col("æº¢é¢åˆ†ä¿è´¹").sum().alias("Hist_Premium"),
        (pl.col("æº¢é¢å·²å†³") + pl.col("æº¢é¢æœªå†³")).sum().alias("Hist_Claims")
    ]).with_columns([pl.col("ä¸šåŠ¡å¹´åº¦").cast(pl.Int32).alias("Year_Int")]).sort("Year_Int")

    # åŸºå‡†ä¿è´¹
    latest_year = history_df["Year_Int"].max()
    current_baseline_premium = history_df.filter(pl.col("Year_Int") == latest_year)["Hist_Premium"][0]
    if current_baseline_premium < 1000:
        current_baseline_premium = history_df.tail(3)["Hist_Premium"].mean()

    # é”å®šè¿‘äº”å¹´
    start_year = latest_year - 4
    period_df = history_df.filter(pl.col("Year_Int") >= start_year)
    
    print(f"## ğŸ’£ æç«¯å‹åŠ›æµ‹è¯•ï¼š2023å¹´ 'æ±¾é£æ½‡æ²³' èµ”æ¡ˆ")
    print(f"**å†²å‡»åŠ›åº¦**: é¢å¤–å¢åŠ èµ”æ¬¾ **{DELTA_LOSS/10000:,.2f}** ä¸‡å…ƒ")
    print(f"**å—å½±å“å¹´ä»½**: 2023 (è¿‘äº”å¹´ä¸šç»©æ ¸å¿ƒåŒº)\n")

    # -------------------------------------------------------------------------
    # 3. æ„å»ºå¯¹æ¯”æ•°æ®é›† (Normal vs Stressed)
    # -------------------------------------------------------------------------
    
    # A. æ­£å¸¸åœºæ™¯
    df_normal = period_df.clone()
    
    # B. å‹åŠ›åœºæ™¯ (æ³¨å…¥æ¯’è¯)
    # æ‰¾åˆ° 2023 å¹´çš„è¡Œï¼Œå¢åŠ èµ”æ¬¾
    df_stressed = period_df.with_columns([
        pl.when(pl.col("Year_Int") == STRESS_YEAR)
          .then(pl.col("Hist_Claims") + DELTA_LOSS)
          .otherwise(pl.col("Hist_Claims")).alias("Hist_Claims")
    ])
    
    # -------------------------------------------------------------------------
    # 4. è®¡ç®—æ ¸å‡½æ•° (Trended + LPC)
    # -------------------------------------------------------------------------
    def calc_metrics(df_input, label):
        # è¶‹åŠ¿åŒ–
        df = df_input.with_columns([
            (pl.col("Hist_Claims") * (pl.lit(1 + TREND_RATE).pow(pl.lit(latest_year) - pl.col("Year_Int")))).alias("Trended_Claims")
        ])
        # èµ”ä»˜ç‡
        df = df.with_columns([
            (pl.col("Trended_Claims") / pl.lit(current_baseline_premium)).alias("Sim_Gross_LR")
        ])
        # LPC
        df = df.with_columns([
            (pl.min_horizontal(pl.max_horizontal(pl.col("Sim_Gross_LR") - LPC_TRIGGER, pl.lit(0.0)), pl.lit(LPC_RANGE)) * LPC_SHARE).alias("LPC_Recovery")
        ])
        df = df.with_columns([
            (pl.col("Sim_Gross_LR") - pl.col("LPC_Recovery")).alias("Sim_Net_LR")
        ])
        
        # æŒ‡æ ‡æå–
        avg_net_lr = df["Sim_Net_LR"].mean() # EL
        
        # é‡ç‚¹å…³æ³¨ 2023 å¹´çš„è¡¨ç°
        year_2023 = df.filter(pl.col("Year_Int") == 2023)
        if year_2023.height > 0:
            lr_2023_gross = year_2023["Sim_Gross_LR"][0]
            lr_2023_net = year_2023["Sim_Net_LR"][0]
        else:
            lr_2023_gross = 0
            lr_2023_net = 0
            
        lpc_hit_cap = lr_2023_gross > LPC_CAP # æ˜¯å¦å‡»ç©¿ä¸Šé™
        
        # èµ„æœ¬å ç”¨
        max_net_lr = df["Sim_Net_LR"].max()
        capital_req = max(max_net_lr - 1.0, 0.35)
        
        return {
            "åœºæ™¯": label,
            "2023 Gross LR": lr_2023_gross,
            "2023 Net LR": lr_2023_net,
            "LPCçŠ¶æ€": "ğŸ”´ å‡»ç©¿ä¸Šé™" if lpc_hit_cap else ("ğŸŸ¡ è§¦å‘åˆ†æ‹…" if lr_2023_gross > LPC_TRIGGER else "ğŸŸ¢ æœªè§¦å‘"),
            "äº”å¹´å¹³å‡èµ”ä»˜ç‡ (EL)": avg_net_lr,
            "èµ„æœ¬å ç”¨ (Capital)": capital_req
        }

    # -------------------------------------------------------------------------
    # 5. æ‰§è¡Œå¯¹æ¯”
    # -------------------------------------------------------------------------
    res_normal = calc_metrics(df_normal, "1. ç°çŠ¶ (Current)")
    res_stress = calc_metrics(df_stressed, "2. å‹åŠ› (Stressed)")
    
    # -------------------------------------------------------------------------
    # 6. è¾“å‡ºç»“æœ
    # -------------------------------------------------------------------------
    res_df = pl.DataFrame([res_normal, res_stress])
    
    display_df = res_df.select([
        pl.col("åœºæ™¯"),
        (pl.col("2023 Gross LR")*100).round(1).alias("23å¹´æ¯›èµ”ä»˜(%)"),
        pl.col("LPCçŠ¶æ€"),
        (pl.col("2023 Net LR")*100).round(1).alias("23å¹´å‡€èµ”ä»˜(%)"),
        (pl.col("äº”å¹´å¹³å‡èµ”ä»˜ç‡ (EL)")*100).round(1).alias("äº”å¹´EL(%)"),
        (pl.col("èµ„æœ¬å ç”¨ (Capital)")*100).round(1).alias("èµ„æœ¬å ç”¨(%)")
    ])
    
    print("### ğŸ“Š å‹åŠ›æµ‹è¯•å¯¹æ¯”è¡¨")
    print_markdown_table(display_df)
    
    # -------------------------------------------------------------------------
    # 7. æŸå®³è¯„ä¼°
    # -------------------------------------------------------------------------
    cap_increase = res_stress["èµ„æœ¬å ç”¨ (Capital)"] - res_normal["èµ„æœ¬å ç”¨ (Capital)"]
    el_increase = res_stress["äº”å¹´å¹³å‡èµ”ä»˜ç‡ (EL)"] - res_normal["äº”å¹´å¹³å‡èµ”ä»˜ç‡ (EL)"]
    
    # è®¡ç®—å¯¹è´¹ç‡çš„å½±å“
    # Impact = Delta_EL + Delta_RiskLoad
    rate_impact = el_increase + (cap_increase * 0.15)
    
    print("\n### ğŸš‘ æŸå®³è¯„ä¼°æŠ¥å‘Š (Damage Assessment)")
    if res_stress["LPCçŠ¶æ€"] == "ğŸ”´ å‡»ç©¿ä¸Šé™":
        print(f"âš ï¸ **LPC é˜²çº¿å¤±å®ˆï¼**")
        print(f"- 2023å¹´çš„æ¨¡æ‹Ÿæ¯›èµ”ä»˜ç‡è¾¾åˆ°äº† **{res_stress['2023 Gross LR']*100:.1f}%**ï¼Œå‡»ç©¿äº† 120% çš„ä¸Šé™ã€‚")
        print("- è¿™æ„å‘³ç€ï¼šè¶…è¿‡ 120% çš„éƒ¨åˆ†ï¼Œå†ä¿äººæ˜¯ **100% å…¨é¢èµ”ä»˜** çš„ã€‚")
    
    print(f"\n**è°ˆåˆ¤ç­¹ç æŸå¤±ä¼°ç®—**ï¼š")
    print(f"1. **EL æ¶åŒ–**: +{el_increase*100:.2f}% (æˆæœ¬ä¸Šå‡)")
    print(f"2. **èµ„æœ¬å ç”¨æ¶åŒ–**: +{cap_increase*100:.2f}% (ä» {res_normal['èµ„æœ¬å ç”¨ (Capital)']*100:.1f}% é£™å‡è‡³ {res_stress['èµ„æœ¬å ç”¨ (Capital)']*100:.1f}%)")
    print(f"3. **æ€»è´¹ç‡å†²å‡»**: **+{rate_impact*100:.2f}%**")
    
    print(f"\n**ğŸ’¡ åº”å¯¹å»ºè®®**ï¼š")
    if rate_impact > 0.05:
        print("ğŸ”´ **å±€é¢æåº¦è¢«åŠ¨**ã€‚å¦‚æœæ­¤æ¡ˆåå®ï¼Œ'é™ä»·/æ¶¨PC' å‡ ä¹ä¸å¯èƒ½ã€‚")
        print("ğŸ‘‰ **å”¯ä¸€å‡ºè·¯**ï¼šå¿…é¡»æ­»ä¿ 'å¶å‘æ€§å·¨ç¾' çš„å®šæ€§ï¼Œè¦æ±‚å°†æ­¤æ¡ˆå‰”é™¤å‡ºå®šä»·æ¨¡å‹ï¼ˆOutlier Treatmentï¼‰ï¼Œæˆ–è€…åˆ†æ‘Šåˆ° 20 å¹´å»æ¶ˆåŒ–ï¼Œç»ä¸èƒ½è®©å®ƒç›´æ¥æ±¡æŸ“è¿‘ 5 å¹´çš„å¹³å‡å€¼ã€‚")
    else:
        print("ğŸŸ¡ **å°šå¯æ‰¿å—**ã€‚è™½ç„¶æœ‰å†²å‡»ï¼Œä½†å¾—ç›Šäºä¿è´¹åŸºæ•°å¤§ï¼Œå¹¶æœªé€ æˆæ¯ç­æ€§æ‰“å‡»ã€‚ä»æœ‰è°ˆåˆ¤ä½™åœ°ã€‚")

    return None

dashboard.set_update_function(update_dashboard)
dashboard.show()
```
